## Importation des bibliothèques

## Chargement des données

```python
# collecter tous les fichiers CSV dans le dossier Data
files = glob.glob(root_path + r'\notebook\Data\*.csv')
```

## Prétraitement des données

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['gaze_direction'].value_counts().plot(kind='bar')
plt.title('Distribution des détections du regard')
plt.xlabel('Détection du regard')
plt.ylabel('Nombre d\'occurrences')
plt.show()
```

```python
# On va appliquer get_dummies sur les colonnes catégorielles
df_m = pd.get_dummies(df, columns=['head_pose', 'gaze_direction'], drop_first=True)
df_m.info()
```

```python
# On melange les lignes du DataFrame
df_m = df_m.sample(frac=1).reset_index(drop=True)
```

```python
df_m
```

```python
df_m['label'].value_counts().plot(kind='bar')
plt.title('Distribution des labels')
plt.xlabel('Label')
plt.ylabel('Nombre d\'occurrences')
plt.show()
```

## Modélisation et évaluation

```python
# On va diviser le DataFrame en ensembles d'entraînement et de test
from sklearn.model_selection import train_test_split
X = df_m.drop('label', axis=1)
y = df_m['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# On va essayer plusieurs modèles de classification
# Regression logistique
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# La matrice de confusion pour le modèle de la regression logistique
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4,4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=logreg.classes_, yticklabels=logreg.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# Om essaye le modele dicision tree
from sklearn.model_selection import RandomizedSearchCV

from sklearn.tree import DecisionTreeClassifier

param_dist = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}
dt = DecisionTreeClassifier()

random_search_dt = RandomizedSearchCV(estimator = dt, param_distributions=param_dist, n_iter=200, cv=5, random_state=42)
random_search_dt.fit(X_train, y_train)
print("Best parameters found: ", random_search_dt.best_params_)
```

```python
dt = random_search_dt.best_estimator_
y_pred_dt = dt.predict(X_test)
print(classification_report(y_test, y_pred_dt))
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
```

```python
# La matrice de confusion pour le modèle Decision Tree
conf_matrix = confusion_matrix(y_test, y_pred_dt)
plt.figure(figsize=(4,4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=dt.classes_, yticklabels=dt.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

param_grid_rf = {
    'n_estimators': [100, 150, 200],
    'criterion': ['gini', 'entropy'],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestClassifier(random_state=0)
random_search_rf = RandomizedSearchCV(estimator=rf, param_distributions=param_grid_rf, 
                                   n_iter=10, cv=5, n_jobs=-1, verbose=2, random_state=42)
random_search_rf.fit(X_train, y_train)
print("Best Hyperparameters:", random_search_rf.best_params_)
```

```python
rf = random_search_rf.best_estimator_
y_pred_rf = rf.predict(X_test)
print(classification_report(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
# La matrice de confusion pour le modèle Random Forest
conf_matrix = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(4,4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=rf.classes_, yticklabels=rf.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# On va faire la meme chose pour XGBoost
from xgboost import XGBClassifier
param_grid_xgb = {
    'n_estimators': [100, 150, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0]
}
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=0)
random_search_xgb = RandomizedSearchCV(estimator=xgb, param_distributions=param_grid_xgb, 
                                   n_iter=10, cv=5, n_jobs=-1, verbose=2, random_state=42)
random_search_xgb.fit(X_train, y_train)
print("Best Hyperparameters:", random_search_xgb.best_params_)
```

```python
# Voir l'accuracy du modèle XGBoost avec les meilleurs hyperparamètres
xgb = random_search_xgb.best_estimator_
y_pred_xgb = xgb.predict(X_test)
print(classification_report(y_test, y_pred_xgb))
```

```python
# L'accuracy du modèle XGBoost
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
```

```python
# On trace la matrice de confusion pour le modèle XGBoost
conf_matrix = confusion_matrix(y_test, y_pred_xgb)
plt.figure(figsize=(4,4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=xgb.classes_, yticklabels=xgb.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
