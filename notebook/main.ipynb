```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import seaborn as sns
```

```python
root_path = os.path.abspath(os.path.join(os.getcwd(), '..'))
```

```python
# collecter tous les fichiers CSV dans le dossier Data
files = glob.glob(root_path + r'\notebook\Data\*.csv')
```

```python
# concaténer tous les fichiers CSV en un seul DataFrame
df= pd.concat([pd.read_csv(file) for file in files], ignore_index=True)
df.head(5)
```

```python
# On va remplacer les valeurs manquantes dans head_pose par 'unknown'
df['head_pose'].fillna('unknown', inplace=True)
```

```python
df['head_pose'].value_counts().plot(kind='bar')
plt.title('Distribution des positions de la tête')
plt.xlabel('Position de la tête')
plt.ylabel('Nombre d\'occurrences')
plt.show()
```

```python
# On va remplir les valeurs manquantes de la colonne 'gaze_direction' Par 'unknown'
df['gaze_direction'].fillna('unknown', inplace=True)
```

```python
df['gaze_direction'].value_counts().plot(kind='bar')
plt.title('Distribution des détections du regard')
plt.xlabel('Détection du regard')
plt.ylabel('Nombre d\'occurrences')
plt.show()
```

```python
# On va appliquer get_dummies sur les colonnes catégorielles
df_m = pd.get_dummies(df, columns=['head_pose', 'gaze_direction'], drop_first=True)
df_m.info()
```

```python
# On melange les lignes du DataFrame
df_m = df_m.sample(frac=1).reset_index(drop=True)
```

```python
df_m
```

```python
df_m['label'].value_counts().plot(kind='bar')
plt.title('Distribution des labels')
plt.xlabel('Label')
plt.ylabel('Nombre d\'occurrences')
plt.show()
```

```python
# On va diviser le DataFrame en ensembles d'entraînement et de test
from sklearn.model_selection import train_test_split
X = df_m.drop('label', axis=1)
y = df_m['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# On va essayer plusieurs modèles de classification
# Regression logistique
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

param_grid_rf = {
    'n_estimators': [100, 150, 200],
    'criterion': ['gini', 'entropy'],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestClassifier(random_state=0)
random_search_rf = RandomizedSearchCV(estimator=rf, param_distributions=param_grid_rf, 
                                   n_iter=10, cv=5, n_jobs=-1, verbose=2, random_state=42)
random_search_rf.fit(X_train, y_train)
print("Best Hyperparameters:", random_search_rf.best_params_)
```

```python
rf = random_search_rf.best_estimator_
y_pred_rf = rf.predict(X_test)
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
```
